```python
import pandas as pd
data=[]
file_path = '서울임대료.csv'  # CSV 파일 경로
df = pd.read_csv(file_path, header=None)  # CSV 파일을 데이터프레임으로 읽기

# 분할하려는 열 선택 (예: A 열)
column_number = 0  # 분할하려는 열의 열 번호 (0부터 시작)
delimiter = '|'  # 원하는 구분자 (예: 쉼표)

# 선택한 열의 모든 셀에 대해 구분자로 데이터 분할
df[column_number] = df[column_number].apply(lambda x: x.split(delimiter))

# 분할된 데이터 출력 (선택 사항)
for row in df.itertuples(index=False):
    
    data.append(row[column_number])

```

```python
df = pd.DataFrame(data[1:], columns=data[0])
```
