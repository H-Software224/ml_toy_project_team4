```python
from folium.plugins import MarkerCluster, HeatMap
import requests
import json
import folium
import pandas as pd

r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
import pandas as pd
from pprint import pprint

model_prediction = pd.read_csv('/content/drive/MyDrive/오늘은 내가 탕후루 요리사/예측값/8.5.18.csv')
pprint(model_prediction)
```

```python
tanghuru_stores = pd.read_csv('/content/drive/MyDrive/오늘은 내가 탕후루 요리사/원본 데이터/tanghulu.csv')
pprint(tanghuru_stores)
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    tiles = 'openstreetmap',
    zoom_start=11
)

folium.GeoJson(
    seoul_geo,
    name='지역구',
).add_to(m)

# estimation - heatmap
"""
HeatMap(
    data = model_prediction[['latitude', 'longitude']],
    radius = 35,
).add_to(m)
"""

HeatMap(
    data=model_prediction[['latitude', 'longitude', '예측값']],#.values.tolist(),
    radius=100,
).add_to(m)


# actual locations of tanghuru stores - marker
marker_cluster = MarkerCluster().add_to(m)

for lat, lon in zip(tanghuru_stores['탕후루/latitude'], tanghuru_stores['탕후루/longitude']):
  folium.Marker([lat, lon], icon = folium.Icon(color='black')).add_to(marker_cluster)

m
```
